```python
#In Other variable Nan or missed value are the only players list and remaining are from administrations, Coaches, Umpires etc
players1.Other.unique()
```

```python
#total number of players
players1.Other.isnull().sum()
```

```python
#filling nan with zeros
players1.Other.fillna(0, inplace = True)
```

### Considering only the players who have played atleast one ODI

```python
#dropping all the players who hasnt played a single ODI
players1.dropna(subset = ['BATTING_ODIs_Mat'], inplace = True)
```

### Since the Gender field is not available. We need to extract the gender from the 'major teams' based on the whether it is a mens or womens team

```python
players1['Major teams'].isna().sum()
```

```python
players1.dropna(subset = ['Major teams'], inplace =True)
```

```python
#0 for male and 1 for female
players1['Gender'] = 0
for index, row in players1.iterrows():
    if 'Women' in row['Major teams']:
        players1.loc[index,['Gender']] = 1
```

```python
players1.Gender.unique()
```

### Considering only players who are between 15 to 45 Age

```python
#filtering players only those who are alive and age between 15 and 45
players1 = players1.loc[(players1.Died == 'Alive')  & (players1.Age >= 15) & (players1.Age <= 45) & (players1.Other == 0) ,]
```

```python
#players1['Batting style'].isnull().sum()/len(players)*100
```

```python
#finding the unique batting style
players1['Batting style'].unique()
```

```python
players1['Batting style'].isna().sum()
```

```python
#dropping the entire row if no data is present in batting style
players1 = players1.dropna(subset = ['Batting style'])
```

```python
#Encoding Righthand batsman and lefthand batsmans as 1's and 0's
#players1.loc[players1['Batting style'] == 'Right-hand bat, Right-hand bat','Batting style encoded'] = 1
players1.loc[players1['Batting style'] == 'Right-hand bat','Batting style encoded'] = 1
players1.loc[players1['Batting style'] == 'Left-hand bat','Batting style encoded'] = 0
```

```python
players1 = players1.astype({"Batting style encoded" : "float64"})
```

```python
#Checking no of lef and right hand batsmans
players1['Batting style encoded'].value_counts()
```

```python
#finding the unique bowling style
players1['Bowling style'].unique()
```

```python
players1['Bowling style'].isna().sum()
```

```python
#dropping the entire row if no data is present in bowling style
players1.dropna(subset = ['Bowling style'], inplace = True)
```

```python
#dropping the entire row if the bowling style is unknown
for index, row in players1.iterrows():
    if 'unknown' in row['Bowling style']:
        players1.drop(index, inplace = True)
```

```python
len(players1['Bowling style'].unique())
```

```python
#Since it is a categorical variable we are doing a one hot encoding
zip_dum_df = pd.get_dummies(players1['Bowling style'])
```

```python
zip_dum_df.shape
```

```python
players2 = pd.concat([players1, zip_dum_df], axis=1)
```

```python
#checking the missing values
players2.isnull().sum()
```

```python
#filling all the missed cells with 0's
players2.fillna(0, inplace = True)
```

```python
import matplotlib.pyplot as plt
import scipy
from scipy.cluster.hierarchy import dendrogram,linkage
from sklearn.preprocessing import StandardScaler
```

```python
#standardizing
players4 = players3.apply(lambda x : (x- np.mean(x))/ np.std(x))
```

```python
from sklearn.cluster import KMeans
```

```python
#copy for backup
players5 = players4.copy()
```

```python
# calculating within group SSE for diff cluster solutions
inertias = []
for k in [10,25,50,100,150,200,250,300]:
    km = KMeans(n_clusters = k, n_init = 10)
    inertia = km.fit(players4).inertia_
    inertias.append(inertia)

    
# plotting scree plot
plt.plot([10,25,50,100,150,200,250,300],inertias)
plt.show()
```

```python
km = KMeans(n_clusters = 50, n_init = 10)
km.fit(players4)
```

```python
players_final['group'] = km.labels_
```
